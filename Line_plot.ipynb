A line plot is a type of graph that shows the relationship between two variables by connecting data points with straight lines.
It is mainly used to display trends or changes in data over time or across ordered values.

**In a line plot:**

* The X-axis usually represents time or sequence.  
* The Y-axis represents the measured value.  
* Points are connected with lines to make it easy to see how the value changes.

**Where Line Plots Are Used:**

**1. Time Series Analysis**

Used to show changes over time.
Examples:
* Stock prices   
* Monthly sales
* Website traffic  
* Temprature changes  

**2. Business Analytics**

Companies use line plots to track:
* Revenue growth   
* Customer trends
* Product demand  

**3. Data Science and Machine Learning**

Used in:
* Model performance graphs   
* Trening vs validation loss
* Accuracy over epochs

**4. Scientific Research**

Scientists use line plots for:
* Experimental results
* Climate data
* Population growth

**5. Finance**

Used for:
* Stock market trends  
* Cryptocurrency price changes

**Customizing Properties of line plot:**
1. color (red, green, blue, pink, yellow, etc.)
2. linewidth (1-10)
3. linestyle (dotted, solid, dashed)
4. marker (circle, traingle, square)
5. markersize (0-10)
6. hue (data column)
7. style (data column)
8. palette (palette colors)
9. ci (None)
10. legend (True, False)



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [1,2,3,4,5,6,7]
y = [20,40,30,40,30,50,30]

df = pd.DataFrame({"Days":x , "No. of people":y})
df.head(7)
```

```python
sns.lineplot(x = "Days", y = "No. of people" , data = df)
plt.show()
```

```python
df  = sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(x = "day" , y = "total_bill", data = df)
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, color="red")
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, linewidth=7)
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, style="time")
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, linestyle="dotted")
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, marker="s", markersize=10)
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, palette="pastal")
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, ci=None)
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, legend=True)
plt.show()
```

```python
sns.lineplot(x="day", y="total_bill", data=df, hue="sex", linewidth=7,
             marker="o", markersize=2, color="red",
             linestyle="dotted", ci=None, legend=True)
plt.show()
```
