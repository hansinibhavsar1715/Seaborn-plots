A swarm plot is a categorical scatter plot that shows individual data points without overlapping.

It is like a strip plot, but points are adjusted (spread out) so they don’t overlap, making the distribution clearer.
* X-axis -> categories
* y-axis -> numerical values
* each dot -> one data point

**Why Swarm Plot is Useful**

**1. No Overlapping Points**

Unlike strip plot, points are arranged so every point is visible.

**2. Clear Distribution**

You can easily see:
* density of values
* clusters
* gaps

**3. Better for Small Datasets**

Works best when dataset size is small to medium.

**Where Swarm Plot is Used**

**1. Exploratory Data Analysis (EDA)**
* Understand distribution of data
* Compare categories

**2. Outlier Detection**

Outliers clearly visible hote hain.

**3. Comparing Groups**

Example:
* Salary by department
* Marks by subject
* Tips by day

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=sns.load_dataset("tips")
data.head()
```

```python
sns.swarmplot(x="day", y="total_bill", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=data, color="red")
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", palette="Set2", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", size=7, data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", size=7,marker="*",edgecolor="pink", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", size=7, linewidth=3, linestyle="dotted", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", dodge=True, data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", order=["Thur", "Fri", "Sat", "Sun"], data=data)
```

```python
sns.swarmplot(y="day", x="total_bill", orient="h", data=data)
```

```python
sns.swarmplot(x="day", y="total_bill", alpha=0.5, data=data)
```

```python
sns.swarmplot(
    x="day",
    y="total_bill",
    hue="sex",
    data=data,
    palette="viridis",
    size=6,
    dodge=True,
    alpha=0.8,
    marker="*"
)
plt.show()
```
