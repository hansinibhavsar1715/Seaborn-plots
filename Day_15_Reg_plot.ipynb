A reg plot (regression plot) is used to show the relationship between two variables along with a regression line.

It combines:
* scatter-plot
* best-fit line

This helps you understand how one variable affects another.  

**Why Reg Plot is Useful:**

**1. Relationship Analysis**

Shows whether variables are:
* positively related
* negatively related
* no related

**2. Trend Visualization**

Regression line clearly shows the overall trend.

**3. Prediction Understanding**

Helps in understanding linear relationships for prediction models.

**Where Reg Plot is Used:**

**1. Data Science & Machine Learning**
* Feature relationship analysis
* Before applying regression models

**2. Business Analytics**
* Sales vs Advertising
* Prices vs Demand

**3. Statistics**
* Correlation and regression analysis

**Customizing properties of Reg Plot:**
1. x and y
2. color
3. scatter
4. scatter_kws
5. line_kws
6. ci
7. order
8. logistic
9. lowess
10. robust
11. logx
12. x_jitter and y_jitter
13. ax
14. truncate









```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
data.head(10)
```

```python
sns.regplot(x="total_bill", y="tip", data=data)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, color="red")
```

```python
sns.regplot(x="total_bill", y="tip", data=data, scatter=False)
```

```python
sns.regplot(
    x="total_bill",
    y="tip",
    data=data,
    scatter_kws={"s": 50, "alpha": 0.6}
)
```

```python
sns.regplot(
    x="total_bill",
    y="tip",
    data=data,
    line_kws={"color": "blue", "linewidth": 2}
)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, ci=None)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, order=2)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, logistic=False)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, lowess=True)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, robust=True)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, logx=True)
```

```python
sns.regplot(x="total_bill", y="tip", data=data, x_jitter=0.5)
```

```python
sns.regplot(
    x="total_bill",
    y="tip",
    data=data,
    scatter_kws={"s": 40, "alpha": 0.6},
    line_kws={"color": "red"},
    ci=95,
    order=1
)
plt.show()
```
