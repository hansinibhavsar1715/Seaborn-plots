A scatter plot is a graph used to show the relationship between two numerical variables.

Each point represents one data observation:

X-axis → one variable

Y-axis → another variable

So every dot = (x, y) value.

**Why Scatter Plot is Useful:**

**1. Shows Relationship (Correlation**)

You can see if variables are:

positively related 📈 (both increase)

negatively related 📉 (one increases, other decreases)

not related

**2. Detects Patterns**

Helps identify:

trends

clusters

patterns

**3. Finds Outliers**

Unusual data points are easy to spot.

**Where Scatter Plot is Used:**

**1. Data Science & Machine Learning**

Feature relationship analysis

Before applying regression models

**2. Business Analytics**

Advertising vs sales

Price vs demand

**3. Scientific Research**

Experiments and observations

**Customizing Properties of Scatter Plot:**
1. x and y
2. data
3. hue
4. style
5. size
6. color
7. palette
8. markers
9. aplha
10. s - marker size (fixed)(100)
11. hue_order
12. size_order
13. legend
14. ax - axis controls


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = sns.load_dataset("exercise")
df.head()
```

```python
sns.scatterplot(x="time", y="pulse", data=df)
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, hue="kind")
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, palette="GnBu")
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, size="diet")
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, markers="size")
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df,markers="*")
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, alpha=0.5)
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, alpha=0.5, hue="kind", hue_order=["rest","walking","running"])
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, alpha=0.5, size="diet", size_order=["low fat","no fat"])
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, alpha=0.5, legend = True)
plt.show()
```

```python
sns.scatterplot(x="time",
                y="pulse",
                data=df,
                hue="kind",
                palette="viridis",
                markers="x",
                alpha=0.8,
                legend = True,
                size = "diet",
                s=100)
plt.show()
```
