A FacetGrid is a Seaborn tool used to create a grid of multiple plots based on categories.

Instead of showing everything in one plot, FacetGrid splits the data into multiple small plots (subplots) based on categorical variables.

Each small plot represents a subset of the data.

**Why FacetGrid is Used**
**1. Compare Categories Easily**

You can compare how data behaves across:
* different genders
* different regions
* different classes

**2. Better Visualization**

Instead of cluttered graphs, data is divided into clean, separate panels.

**3. Advanced EDA**

Very useful in Exploratory Data Analysis when dealing with multiple variables.

**Customizing properties of FacetGrid plot:**
1. row
2. col
3. hue
4. col_wrap - wraps column into multiple rows (1 and above)
5. height
6. aspect
7. sharex and sharey
8. row_order and col_order
9. hue_order
10. palette
11. legend
12. map()
13. map_dataframe()

```python
sns.FacetGrid(data=df, col="time", hue="sex")
plt.show()
```

```python
sns.FacetGrid(data=df, col="time", col_wrap=2)
plt.show()
```
