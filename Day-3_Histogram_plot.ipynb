A histogram is a graph used to show the distribution of numerical (continuous) data using bars.

It divides data into intervals (called bins) and shows how many values fall into each interval.

X-axis → value ranges (bins)

Y-axis → frequency (number of values)

**Example**

Data:

[10, 12, 15, 18, 20, 22, 25]

Histogram will group values like:

10–15 → count

15–20 → count

20–25 → count

Each group is shown as a bar.

**Why Histogram is Used**

**1. To Understand Data Distribution**

It helps you see:

where most values are concentrated

how data is spread

**2. To Identify Shape of Data**

You can detect:

normal distribution

skewness (left or right)

uniform distribution

**3. To Detect Outliers**

Extreme values become easy to spot.

**Where Histogram is Used**

**1. Exploratory Data Analysis (EDA)**

To understand data before building models

**2. Data Science and Machine Learning**

To analyze feature distributions

**3. Statistics**

To study frequency distribution

**Customizing Properties of Histogram:**
1. bins - control number of bars
2. color
3. hue
4. kde
5. alpha
6. multiple
7. stat - defines what y-ais shows
8. binwidth / binrange

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = sns.load_dataset("tips")
df.head()
```

```python
sns.histplot(x="total_bill", data=df)
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, bins = 10)
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, color="green")
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, hue="sex", palette="spring")
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, discrete="False")
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, kde=True)
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, hue="sex", alpha=0.5)
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, linestyle="dotted")
plt.show()
```

```python
sns.histplot(x="total_bill", data=df, edgecolor="red", color="hotpink", linestyle="--")
plt.show()
```

```python
sns.histplot(x="total_bill",
             data=df,
             hue="sex",
             bins=15,
             binwidth=1.5,
             palette="spring",
             discrete=False,
             alpha=0.5,
             linestyle="dotted")
plt.show()
```
