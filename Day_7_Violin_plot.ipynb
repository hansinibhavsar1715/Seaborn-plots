A violin plot is used to show the distribution of numerical data across categories, combining:

Box plot (median, quartiles)
KDE (density curve) (shape of data)

It looks like a violin shape, where:

Width → density (more data = wider)

Center line → median

Shape → distribution

**Why Violin Plot is Useful:**

**1. Shows Full Distribution**

Unlike box plot, it shows how data is spread (shape).

**2. Detects Patterns**

You can see:

multiple peaks

skewness

clusters

**3. Better than Box Plot (sometimes**)

Box plot → summary

Violin plot → summary + distribution

**Where Violin Plot is Used:**

**1. Exploratory Data Analysis (EDA)**

Understanding distribution across categories

**2. Data Science / ML**

Feature distribution comparison

**3. Research & Statistics**

Comparing groups

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df= sns.load_dataset("tips")
df.head(7)
```

```python
sns.violinplot(x="total_bill", data=df)
plt.show()
```

```python
sns.violinplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df)
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, hue="day")
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, color= "blue")
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, palette="viridis")
plt.show()
```

```python
sns.violinplot(x="tip", y="sex", data=df)
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, hue="day", split=True, saturation=0.5)
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, order=["Female","Male"], linewidth=3)
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, inner="quartile",)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", palette="Set2", data=df)
plt.show()
```

```python
sns.violinplot(x="sex",
               y="tip",
               data=df,
               hue="day",
               color="blue",
               palette="viridis",
               split=False,
               linewidth=3)
plt.show()
```
