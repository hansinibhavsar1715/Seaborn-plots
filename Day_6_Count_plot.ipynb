A count plot is used to show the frequency (count) of observations in each category.

It automatically counts how many times each category appears and displays it as bars.

X-axis → categories
Y-axis → count (number of occurrences)

**Why Count Plot is Useful:**

**1. Quick Frequency Analysis**

You can instantly see:
which category is most common
which is least common

**2. No Manual Calculation**

No need to calculate counts manually — it does it automatically.

**3. Easy Comparison**

Helps compare categories visually.

**Where Count Plot is Used:**

**1. EDA (Exploratory Data Analysis)**

Understanding categorical data

**2. Business Analytics**

Customer segments
Product categories

**3. Surveys & Research**

Responses distribution

**Customizing Properties of Count Plot:**
1. x or y
2. data
3. hue
4. color
5. palette
6. order
7. hue_order
8. horizontal plot
9. dodge
10. linewidth
11. edgecolor
12. ax


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset("tips")
df.head()
```

```python
#sns.countplot(data=df, x="tips")
#sns.countplot(data=df, x="sex")
sns.countplot(data=df, y="tip")
plt.show()
```

```python
#sns.countplot(data=df, x="sex", y="tip")
sns.countplot(data=df, x="sex", hue="smoker")
plt.show()
```

```python
sns.countplot(data=df, x="sex", hue="smoker", palette="spring",)
plt.show()
```

```python
sns.countplot(data=df, x="sex", hue="smoker", color="hotpink", saturation=0.2)
plt.show()
```

```python
#sns.countplot(data=df, x="sex", hue="smoker", facecolor=[0,0,0,1])
#sns.countplot(data=df, x="sex", hue="smoker", facecolor=[1,0,0,1])
#sns.countplot(data=df, x="sex", hue="smoker", facecolor=[0,1,0,1])
sns.countplot(data=df, x="sex", hue="smoker", facecolor=[0,0,1,1])
plt.show()
```

```python
sns.countplot(data=df, x="sex", hue="smoker", linewidth= 2, edgecolor="green")
plt.show()
```

```python
sns.countplot(data=df,
              x="sex",
              hue="smoker",
              palette="viridis",
              color="hotpink",
              saturation=0.2,
              edgecolor="green")
plt.show()
```
