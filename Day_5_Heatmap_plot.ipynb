A heatmap is a data visualization that shows values using colors.

Instead of numbers, colors represent values

Dark / intense colors → higher values

Light colors → lower values

It is usually displayed in a matrix (grid format).

**Why Heatmap is Useful:**

**1. Easy Pattern Recognition**

Colors make it easy to quickly see:

high values

low values

patterns

**2. Correlation Analysis**

Most commonly used to show correlation between variables.

**3. Large Data Simplification**

Big datasets become easier to understand using colors instead of numbers.

**Where Heatmap is Used:**

**1. Data Science & Machine Learning**

Correlation matrix

Feature relationships

**2. Business Analytics**

Sales performance across regions

Customer activity

**3. Research & Statistics**

Experimental data analysis

**Customizing properties of Heatmap Plot:**
1. data - matrix data input
2. cmap - color map
3. annot - show values inside cells (True, False)
4. fmt - format of numbers
5. vmin and vmax - set value range
6. center - center value for diverging colors
7. linewidhts - line thickness between cells
8. linecolor - grid line color
9. square - make cells square
10. xticklabel / yticklabel - show/hide labels
11. cbar - show/hide color bar
12. cbar_kws - customize color bar
13. mask - hide some values
14. ax - axis control


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data= np.random.randint(low=1, high=100, size=(10,10))
print(data)
```

```python
sns.heatmap(data=data)
plt.show()
```

```python
sns.heatmap(data=data, cmap="coolwarm")
plt.show()
```

```python
sns.heatmap(data=data, cbar=False)
plt.show()
```

```python
sns.heatmap(data, annot=True, fmt=".2f")
plt.show()
```

```python
sns.heatmap(data, vmin=-1, vmax=1)
plt.show()
```

```python
sns.heatmap(data=data, cmap="GnBu", alpha=0.5)
plt.show()
```

```python
sns.heatmap(data, square=True)
plt.show()
```

```python
sns.heatmap(data=data, linecolor= "yellow", linewidth= "2")
plt.show()
```

```python
sns.heatmap(data=data, annot=True)
plt.show()
```

```python
sns.heatmap(data=data, xticklabels=False, yticklabels=False)
plt.show()
```

```python
import numpy as np
mask = np.triu(data)

sns.heatmap(data, mask=mask)
plt.show()
```

```python
sns.heatmap(data=data,
            cmap="GnBu",
            cbar=False,
            alpha=0.5,
            linecolor="yellow",
            linewidths="2",
            annot=True,
            square=True,
            xticklabels=True,
            yticklabels=True)
plt.show()
```
