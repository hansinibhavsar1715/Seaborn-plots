A box plot (box-and-whisker plot) is used to show the statistical summary of data distribution.

It summarizes data using:
* Minimum
* Q1 (25%)  
* Median (50%)
* Q3 (75%)  
* Maximum
* Outliers  

Components of a Box Plot:
* Box → from Q1 to Q3 (Interquartile Range - IQR)
* Line inside box → Median
* Whiskers → extend to min and max values (excluding outliers)
* Dots outside → Outliers

Why Box Plot is Useful:

1. Shows Data Summary

Gives a quick overview of:
* central tendency
* spread  

2. Detects Outliers
* Outliers are clearly visible as points.

3. Compare Categories
* Easy to compare distributions across groups.

Where Box Plot is Used:

1. Exploratory Data Analysis (EDA)

Understanding data distribution

2. Data Science / ML

Feature analysis
Outlier detection

3. Business Analytics

Comparing performance across categories

Customizing Properties of Box Plot:
1. x and y
2. data
3. hue
4. palette
5. color
6. linewidth
7. width
8. showfliers - show/hide outliers
9. fliersize - size of outliers points
10. whis - control wisker length
11. order
12. hue_order
13. orient
14. ax

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
sns.boxplot(x="tip", data=df)
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df)
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, hue="smoker")
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, palette="viridis")
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, color=[1,0,0,1])
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, showfliers=False)
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, whis=1.5)
plt.show()
```

```python
sns.boxplot(x="sex", y="tip", data=df, linewidth=3)
plt.show()
```

```python
#sns.boxplot(data=df, orient="h")
sns.boxplot(data=df["tip"], orient="u")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="tip", data=df)
plt.show()
```

```python
sns.boxplot(x="sex",
            y="tip",
            data=df,
            hue="smoker",
            palette="viridis",
            color=[1,0,0,1],
            linewidth=2)
plt.show()
```

```python
sns.boxplot(
    x="sex",
    y="tip",
    hue="smoker",
    data=df,
    palette="Set2",
    color = [0,1,0,1],
    linewidth = 2,
    fliersize = 3,
    whis = 1.5,
    order = ["Male", "Female"],
    hue_order = ["Yes", "No"],
    showfliers=True,
    width=0.6
)

plt.title("Customized Box Plot")
plt.show()
```
