A strip plot is a categorical scatter plot used to show the distribution of data points across different categories.

* X-axis → categorical variable (like product, gender, class)  
* Y-axis → numerical values
* Each point represents an individual observation

**Where Strip Plot is Used**

**1. Data Distribution Analysis**

Used to see how values are spread within each category.

Example:
* Salary disdributaion by departments
* Marks distribution by subjects  

**2. Detecting Outliers**

Since all points are visible, you can easily spot:
* Extremly high or low values

**3. Small to Medium Datasets**

Best when dataset size is not too large (otherwise points overlap).

**4. EDA (Exploratory Data Analysis)**

Used to:
* Undertand row data
* Comapare distribuations across categories

**Customizing properties of Strip Plot:**
1. hue
2. palette
3. color
4. size
5. edgecolor
6. linewidth
7. marker
8. jitter - Adds random noise to avoid overlapping(True, float_value)
9. dodge - separates points when using hue (True, False)
10. orient
11. order
12. hue_order
13. alpha (0.1 and above)



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset("tips")
df.head(10)
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", color="red")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="Set2")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", size=8)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", edgecolor="red",)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", linewidth=1, edgecolor="pink")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", marker="s")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", orient="h")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", order=["Sun","Sat","Fri","Thur"])
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", jitter=0.2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", dodge=True)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", alpha=0.5)
plt.show()
```

```python
sns.stripplot(
    x="day",
    y="total_bill",
    hue="sex",
    data=df,
    color="red",
    size=7,
    jitter=True,
    dodge=True,
    alpha=0.7,
    edgecolor="black",
    marker="*",
    orient="v"
)
plt.show()
```
