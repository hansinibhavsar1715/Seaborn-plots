A point plot is used to show summary statistics (usually mean) of a numerical variable across different categories.

Instead of showing all data points, it shows:
* points = average values (mean)
* lines = trend across categories
* error bars = variability (confidence interval)  

**Why Point Plot is Useful:**

**1. Compare Averages**

Helps compare mean values across categories clearly.

**2. Show Trends**

Lines connect points, making it easy to see:
* increase/decrease
* patterns

**3. Cleaner than Bar Plot**

Bar plot → filled bars
Point plot → minimal & clearer (less clutter).

**Where Point Plot is Used:**

**1. Business Analysis**
* Average sales per category
* Average rating per product

**2. Data Science / EDA**
* Compare group statistics
* Understand category-wise trends

**3. Experiments & Research**
* Comparing results across groups

**Customizing properties of Point Plot:**
1. x and y
2. hue
3. estimator
4. ci - confidence interval control
5. erroebar - controls error bars
6. color
7. palette
8. markers
9. linestyles
10. sacle - controls size of marker and line
11. order
12. hue_order
13. orient
14. dodge
15. join - connects points with line or not
16. capsize - error bar caps
17. errwidth - error bar thickness
18. ax





```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
data.head(7)
```

```python
sns.pointplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex", data=data)
```

```python
import numpy as np
sns.pointplot(x="day", y="total_bill", data=data, estimator=np.mean)
```

```python
sns.pointplot(x="day", y="total_bill", data=data, ci=None)
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex", palette="Set2", data=data)
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex", markers=["o", "s"], data=data)
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex", linestyles=["-", "--"], data=data)
```

```python
sns.pointplot(x="day", y="total_bill", data=data, scale=1.5)
```

```python
sns.pointplot(y="day", x="total_bill", data=data)
```

```python
sns.pointplot(x="day", y="total_bill", data=data, join=False)
```

```python
sns.pointplot(x="day", y="total_bill", data=data, capsize=0.2)
```

```python
sns.pointplot(
    x="day",
    y="total_bill",
    hue="sex",
    data=data,
    palette="Set2",
    estimator=np.mean,
    markers=["o", "s"],
    linestyles=["-", "--"],
    capsize=0.2
)
plt.show()
```
