A KDE plot is used to visualize the distribution of data as a smooth continuous curve.

Instead of showing bars like a histogram, KDE plot shows a smooth line that represents the probability density of data.
* X-axis -> values  
* Y-axis -> density

**Why KDE Plot is Useful**

**1. Smooth Distribution View**

It gives a clean and smooth understanding of data distribution.

**2. Better than Histogram (sometimes)**

Histogram:
* shows bars

KDE:
* shows smooth curve
* easier to see patterns  

**3. Compare Multiple Distributions**

You can easily compare different categories using multiple curves.

**Where KDE plot is used:**

**1. EDA:**
* Understand data distribution
* check skewness  

**2. Data science and Machine Learning**
* Feature distribuation analysis
* Compare classes

**3. Statistics**
* Probability density estimation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=sns.load_dataset("tips")
data.head(10)
```

```python
sns.kdeplot(data=data, x="total_bill")
```

```python
sns.kdeplot(data=data, x="total_bill", hue="sex")
```

```python
sns.kdeplot(data=data, x="total_bill", fill=True)
```

```python
sns.kdeplot(data=data, x="total_bill", bw_adjust=0.5)
```

```python
sns.kdeplot(data=data, x="total_bill", cut=0)
```

```python
sns.kdeplot(data=data, x="total_bill", clip=(0, 50))
```

```python
sns.kdeplot(data=data, x="total_bill", hue="sex", multiple="stack")
```

```python
sns.kdeplot(data=data, x="total_bill", cumulative=True)
```

```python
sns.kdeplot(data=data, x="total_bill", y="tip")
```

```python
sns.kdeplot(
    data=data,
    x="total_bill",
    hue="sex",
    fill=True,
    palette="Set2",
    bw_adjust=1,
    alpha=0.6
)
plt.show()
```
