A bar plot (bar chart) is a graph used to compare values across different categories.
Each category is represented by a rectangular bar, and the height (or length) of the bar represents the value.

* X-axis → categories (products, cities, departments, etc.)
* Y-axis → numerical values (sales, count, revenue, etc.)    

**Where bar plots are used:**

**1. Comparing Categories**
Bar plots are mainly used when we want to compare different groups or categories.

Example:
* Sales of different products
* Number of students in different courses

**2. Business Analytics**
Companies use bar plots to visualize:

* Product performance
* Revenue by region
* Customer segments

**3. Survey and Research data**
Used to show results like:

* Number of people preferring different brands
* Population distribution across cities

**4. Data Science and Machine Learning**
Bar plots help in:

* Feature importance visualization
* comparing model performance
* Category-wise data analysis

**Customizing properties of barplot:**

1. color("red","green",etc)
2. palette("Set2","viridis",etc)
3. hue("data column")
4. order(["A","B","C"])
5. ci(None)
6. estimator(np.mean, np.sum, np.median,etc)
7. orient("h","v")
8. legend(True, False)
9. saturation(0.1 and above)
10. facecolor([1,1,1,1])
11. edgecolor("0.1 and above")



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x = "who", y="fare", data=df)
plt.show()
```

```python
sns.barplot(x = "who", y="fare", data=df, color="magenta", hue="sex", palette="spring",saturation=0.2, ci=False)
plt.show()
```

```python
sns.barplot(x="fare", y="who", data=df, estimator= np.mean)
plt.show()
```

```python
sns.barplot(x="fare", y="who", data=df, estimator= np.median)
plt.show()
```

```python
sns.barplot(x="fare", y="who", data=df, facecolor=[1,1,1,0], edgecolor="0.2")
plt.show()
```

```python
sns.barplot(x="who", y="fare", data=df, order=["child","women","man"])
plt.show()
```
